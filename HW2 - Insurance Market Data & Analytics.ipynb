```python
enroll_df = pd.read_csv('CPSC_Enrollment_Info_2021_01.csv')
```

```python
states = ['AZ','GA','KY','NV','RI','TX','VT','WV','WI']
filt = (enroll_df['Contract Number'].str.contains('H|E|R')) & (
    enroll_df['Enrollment'] != '*') & (enroll_df['State'].isin(states))
enroll = enroll_df.loc[filt] # .loc Locates Info from 'filt' line of code above
print(enroll.shape)
enroll.head()
```

```python
plan_df = pd.read_excel(
    'Monthly_Report_By_Plan_2021_01.xlsx', skiprows=5, nrows=6969)
filt2 = (plan_df['Contract Number'].str.contains(
    r'[HER]\d{4}', na=False, case=False, regex=True)) & (plan_df['Enrollment '] != '*')
plan = plan_df.loc[filt2]
print(plan.shape)
plan.head()
```

```python
major_df = pd.read_excel('MajorInsuranceOrgs.xlsx')
major = pd.merge(major_df, plan, on='Organization Marketing Name', how='inner')
print(major.shape)
major.head()
```

```python
enroll['Enrollment'] = enroll['Enrollment'].astype(int)
enroll.dtypes
```

```python
major = major.dropna()
major['Enrollment '] = major['Enrollment '].astype(int)
major['Plan ID'] = major['Plan ID'].astype(int)
major.dtypes
```

```python
total = pd.merge(major, enroll, on=['Contract Number'], how='inner')
total
```

```python
total = pd.merge(major, enroll, on=['Contract Number', 'Plan ID'], how='inner')
totalq1 = total.iloc[:, [1, 14, 16]]
print(totalq1.shape)
totalq1.head()
```

```python
grpq1 = totalq1.groupby('State')
dic = {}
for i in states:
    tempq1 = grpq1.get_group(i)
    aggq1 = tempq1.groupby('MajorInsuranceOrgName').agg(
        {'Enrollment': 'sum'}).reset_index().sort_values(by='Enrollment', ascending=False)
    s = sum(aggq1['Enrollment'])
    aggq1['percent_squared'] = (aggq1['Enrollment']/s*100)**2
    hhi = round(sum(aggq1['percent_squared']))
    dic[i] = (aggq1['MajorInsuranceOrgName'].iloc[0], hhi)
    
dfq1 = pd.DataFrame(dic).T.reset_index()
dfq1.columns = ['state', 'lion_share_company', 'HHI']
dfq1 = dfq1.sort_values(by='HHI', ascending=False)
dfq1
    
```

```python
top_four = dfq1.iloc[:4, ]
top_four
```

```python
top_four.to_csv('question1.csv', index=False)
```

```python
pbp_df = pd.read_csv('pbp_b16_dental.txt', sep="\t")
print(pbp_df.shape)
pbp_df.head()
```

```python
total2 = total.iloc[:, [1, 2, 3, 14, 16]]
print(total2.shape)
total2.head()
```

```python
column_list = ['pbp_a_hnumber', 'pbp_a_plan_identifier',
               'pbp_b16a_bendesc_yn', 'pbp_b16b_bendesc_yn']

data = pbp_df.loc[:, column_list]
data.columns = ['Contract Number', 'Plan ID', 'preventive', 'comprehensive']
data = data.dropna()
data['preventive'] = 2 - data['preventive']
data['comprehensive'] = 2 - data['comprehensive']
print(data.shape)
data.head()
```

```python
totalq2 = pd.merge(total2, data, on=[
    'Contract Number', 'Plan ID'], how='inner')
print(totalq2.shape)
totalq2.head()
```

```python
grpq2 = totalq2.groupby('State')

dfq2 = pd.DataFrame(columns=['MajorInsuranceOrgName',
                            'Enrollment', 'preventive', 'comprehensive', 'state', 'market_share(%)', 'rank'])
for i in states:
    tempq2 = grpq2.get_group(i)
    aggq2 = tempq2.groupby('MajorInsuranceOrgName').agg(
    {'Enrollment': 'sum', 'preventive':'mean', 'comprehensive': 'mean'}).reset_index().sort_values(
        by='Enrollment', ascending=False)
    aggq2['state'] = i
    s = sum(aggq2['Enrollment'])
    aggq2['market_share(%)'] = aggq2['Enrollment']/s*100
    aggq2 = aggq2.reset_index(drop=True)
    aggq2['rank'] = aggq2.index + 1
    dfq2 = pd.concat([dfq2, aggq2.head()], axis=0, ignore_index=True)
    dfq2
```

```python
dfq2 = dfq2.iloc[:, [4, 6, 0, 5, 2, 3]]
dfq2['market_share(%)'] = dfq2['market_share(%)'].apply(lambda x: round(x, 2))
dfq2['preventive'] = dfq2['preventive'].apply(lambda x: round(x, 2))
dfq2['comprehensive'] = dfq2['comprehensive'].apply(lambda x: round(x, 2))
print(dfq2.shape)
dfq2.head()
```

```python
dfq2.to_csv('question2.csv', index=False)
```

```python
uod3 = pd.read_excel('HEDIS2019.xlsx', sheet_name='hedis_measures', usecols=['CMSContractNumber', 'MeasureCode', 'Rate'])
uod3.columns = ['Contract Number', 'MeasureCode', 'rate']
print(uod3.shape)
uod3.head()
```

```python
uod = uod3.loc[uod3['MeasureCode'] == 'UOD', ['Contract Number', 'rate']]
uod = uod.dropna()
uod = uod.loc[uod['rate'].str.contains('[0-9.-]')]
uod['rate'] = pd.to_numeric(uod['rate'])
print(uod.shape)
uod.head()
```

```python
total3 = pd.merge(total, uod, on='Contract Number', how='inner')
print(total3.shape)
total3.head()
```

```python
totalq3 = total3.iloc[:, [1, 2, 3, 14, 16, 17]]
print(totalq3.shape)
totalq3.head()
```

```python
grpq3 = total3.groupby('State')

dfq3 = pd.DataFrame(
    columns=['MajorInsuranceOrgName', 'Enrollment', 'weighted_average', 'state', 'market_share(%)', 'rank'])

for i in states:
    tempq3 = grpq3.get_group(i)
    aggq3 = tempq3.groupby(['MajorInsuranceOrgName', 'Contract Number']).agg(
        {'Enrollment': 'sum', 'rate': 'mean'}).reset_index().sort_values(by='Enrollment', ascending=False)
    aggq3['weighted_average'] = aggq3['Enrollment'] * aggq3['rate']
    agggrpq3 = aggq3.groupby('MajorInsuranceOrgName').agg(
        {'Enrollment': 'sum', 'weighted_average': 'sum'}).reset_index().sort_values(by='Enrollment', ascending=False)
    agggrpq3['weighted_average'] = agggrpq3['weighted_average'] / \
        agggrpq3['Enrollment'] 
    agggrpq3['weighted_average'] = agggrpq3['weighted_average'].apply(
        lambda x: round(x, 2))
    agggrpq3 = agggrpq3.head(10).sort_values(
        by='weighted_average', ascending=False)
    agggrpq3['state'] = i
    s = sum(agggrpq3['Enrollment'])
    agggrpq3['market_share(%)'] = agggrpq3['Enrollment']/s*100
    agggrpq3 = agggrpq3.reset_index(drop=True)
    agggrpq3['rank'] = agggrpq3.index + 1
    dfq3 = pd.concat([dfq3, agggrpq3], axis=0, ignore_index=True)
    
dfq3['market_share(%)'] = dfq3['market_share(%)'].apply(lambda x: round(x, 2))
dfq3
```

```python
dfq3 = dfq3.iloc[:, [3, 5, 0, 2, 4]]
print(dfq3.shape)
dfq3.head()
```

```python
dfq3.to_csv('question3.csv', index=False)
```
